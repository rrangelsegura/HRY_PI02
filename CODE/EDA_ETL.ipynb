```python

# Definir una función para determinar si se debe incluir una columna
def incluir_columna(nombre_columna):
    # Incluir las columnas que no comienzan con 'Unnamed' son columnas vacias
    return not(nombre_columna.startswith('Unnamed'))


# Definir una función para determinar si se debe incluir una fila
def incluir_fila(fila):
    # Incluir las fila que tienen al menos un dato
    return pd.notna(fila)
```

```python
# Creation a Array with all sources and each source what content it has
sources = [xls_1,xls_2,xls_3,xls_4,xls_5,xls_6,xls_7,xls_8,xls_9,xls_10]
names_sheets = []
df_sources = pd.DataFrame()
for i in sources:
    names = pd.ExcelFile(i, engine='openpyxl').sheet_names
    names_sheets.append(names)
    for e in names_sheets[-1]:
        columns = ((pd.read_excel(i, sheet_name=e, engine='openpyxl', usecols=lambda col: incluir_columna(col)).columns).to_numpy()).tolist()
        linea = {'excel_source':i, 'name sheet': e, 'columns':columns}    
        df_sources = pd.concat([df_sources, pd.DataFrame([linea])], ignore_index=True)
df_sources
```

```python
#Revision de total elementos de cada dataframe y sus faltantes (na) y duplicados, para cada source and their sheets
cantidad_na = []
cantidad_duplicados = []
cantidad_total = []


for row, i in df_sources.iterrows():
    df = pd.read_excel(i[0], sheet_name=i[1], engine='openpyxl', usecols=lambda col: incluir_columna(col))
    #Se decartan filas completamente nulas si fueron cargadas por error
    df = df.dropna(axis=0, how='all')
    cantidad_total.append(((df.shape[0])*(len(df.columns))))
    cantidad_na.append(df.isna().sum().sum())
    cantidad_duplicados.append(df.duplicated().sum())    
```

```python
# CONSTRUIR PARA CORRER NA, DUPLICADOS DE CADA UNO DE LOS DATAFRAME-----------------------------------------------------------------------
#Revision de total elementos de cada dataframe y sus faltantes (na) y duplicados, para cada source and their sheets

cantidad_celdas_na = []
cantidad_rows_duplicados = []
cantidad_celdas_total = []
cantidad_rows_total = []
conteo = 0
sources_df = pd.DataFrame()

for i in sources_list:
    cantidad_rows_total.append(i.shape[0])
    cantidad_celdas_total.append(((i.shape[0])*(len(i.columns))))
    cantidad_celdas_na.append(i.isna().sum().sum())
    cantidad_rows_duplicados.append(i.duplicated().sum())

    nombre_source = sources_names[conteo]
    descripcion_source = description_sources[conteo]
    
    duplicados_porcentage = round((cantidad_rows_duplicados[conteo]/ cantidad_celdas_total[conteo])*100,0)
    na_porcentage = round((cantidad_celdas_na[conteo] / cantidad_celdas_total[conteo])*100,0)

    diccionario = {'nombre_source':nombre_source,
             "cantidad_rows_total":cantidad_rows_total[conteo],
             "cantidad_celdas_total":cantidad_celdas_total[conteo],
             "cantidad_celdas_na":cantidad_celdas_na[conteo],
             "na_porcentage":na_porcentage, 
             "cantidad_rows_duplicados":cantidad_rows_duplicados[conteo],
             "duplicados_porcentage":duplicados_porcentage,
             "descripcion_source":descripcion_source,
             "columns": i.columns.to_numpy().tolist()}
    sources_df = pd.concat([sources_df,pd.DataFrame([diccionario])], ignore_index=True)
    conteo += 1
    
sources_df
```
